```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
with open('Salary_dataset.csv', 'r') as f:
    data = pd.read_csv(f)

print(data.head())
x = 109
```

```python
# split and scale data
train = pd.DataFrame(data['YearsExperience'])
label = pd.DataFrame(data['Salary'])

scaler = StandardScaler()
train = scaler.fit_transform(train)
X_train, X_test, y_train, y_test = train_test_split(train, label, test_size=0.2, random_state=22520109)
y_train = np.ravel(y_train)
y_test = np.ravel(y_test)
print('X_train, y_train shape:', X_train.shape, y_train.shape)
print('X_test, y_test shape:',X_test.shape, y_test.shape)
```

# **SKLEARN MODEL**

```python
# Using sklearn logistic regression model
model = linear_model.LinearRegression(fit_intercept=True)
model.fit(X_train, y_train)
y_predict = np.array(model.predict(X_test))
accuracy = r2_score((y_test), y_predict)
print(f"R2 score for fit intercep = False: {accuracy:.4f}")


# intercept = False -> them bias thu cong
X_train_intercepF = np.c_[X_train, np.ones(X_train.shape[0])]
X_test_intercepF = np.c_[X_test, np.ones(X_test.shape[0])]
model = linear_model.LinearRegression(fit_intercept=False)
model.fit(X_train_intercepF, y_train)
y_predict = np.array(model.predict(X_test_intercepF))
accuracy = r2_score((y_test), y_predict)
print(f"R2 score for fit intercep = True:  {accuracy:.4f}")
```

# **NORMAL EQUATION**

```python
# using normal equation
# add bias
X_train_NE = np.c_[np.ones(X_train.shape[0]), X_train]
X_test_NE = np.c_[np.ones(X_test.shape[0]), X_test]
print(X_train_NE.shape, X_test_NE.shape)

y_train = y_train.reshape(-1, 1)    # reshape y_train to (n, 1)

# normal equation
theta = np.linalg.inv(X_train_NE.T @ X_train_NE) @ X_train_NE.T @ y_train
y_predict = np.array(X_test_NE @ theta)
accuracy = r2_score(y_test, y_predict)
print(f"R2 score for normal equation: {accuracy:.4f}")
```

# **GRADIENT DESCENT**

```python
# using gradient descent
from sklearn.linear_model import SGDRegressor
model = SGDRegressor(
    learning_rate='constant',
    eta0=0.01, 
    max_iter=1000, 
    tol=1e-3, 
    random_state=43, 
    fit_intercept=True
)
y_train = np.ravel(y_train)
model.fit(X_train, y_train)
y_predict = np.array(model.predict(X_test))
accuracy = r2_score(y_test, y_predict)
print(f"R2 score for gradient descent: {accuracy:.4f}")
```
