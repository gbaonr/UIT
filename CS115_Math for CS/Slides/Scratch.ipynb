```python
from matplotlib.image import imread 
import matplotlib.pyplot as plt 
import os 
import numpy as np 
plt.rcParams['figure.figsize'] = [9, 4]
```

```python
os.chdir(r'D:\NguyenBao')
A = imread('flower.jpg')
x = np.mean(A, -1)
```

```python
#show origin
plt.axis('off')
plt.title('Origin', size = 15)
plt.imshow(A)
plt.show()
```

```python
#generate gray img
plt.axis('off')
img = plt.imshow(256-x)
img = img.set_cmap('gray')

plt.title('gray image', size = 15)

```

```python
U, S, VT = np.linalg.svd(x, full_matrices=False)
S = np.diag(S)
```

```python
j = 0 
for r in (5, 20, 100, 500):
    #construct approximate image
    Xapprox = U[:,:r] @ S[0:r,:r] @ VT[:r,:]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(256 - Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S)) / np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.show()

rate = np.cumsum(np.diag(S)) / np.sum(np.diag(S))

for i in (5, 50, 100, 200, 500):
    print(f'Capture rate at sigma {i:}: {rate[i]:.2f}')



```
