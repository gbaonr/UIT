```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# To d liu mu
np.random.seed(0)
X = np.random.rand(10, 1)
X = np.linspace(0, 1, 10)
X = X.reshape(10, 1)

noise = np.random.normal(loc=0, scale=0.03, size=(10, 1))

y = np.sin(1 + X**2) + noise  # Hm s thc t (y ch l v d)

# To cc m hnh v fit vi d liu hun luyn
degrees = np.arange(1, 10)
mse_list = []

print("X shape = ", X.shape)
print("y shape = ", y.shape)
plt.scatter(X, y)

plt.figure(figsize=(12, 6))
for i, degree in enumerate(degrees):
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    
    
    model = LinearRegression()
    model.fit(X_poly, y)
    
    # D on gi tr
    y_pred = model.predict(X_poly)
    
    # o lng  chnh xc bng mean squared error
    mse = mean_squared_error(y, y_pred)
    mse_list.append(mse)
    
    # V biu 
    plt.subplot(3, 3, i + 1)
    plt.scatter(X, y, s=10)
    plt.plot(X, y_pred, color='red')
    plt.title(f'Degree {degree}, MSE: {mse:.2f}')

plt.tight_layout()
plt.show()

# In ra MSE ca tng m hnh
# for degree, mse in zip(degrees, mse_list):
#     print(f"Degree {degree}: MSE = {mse:.2f}")
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# To d liu mu
np.random.seed(0)
X = np.linspace(0, 10, 100)
y = 2 * X**2 - 3 * X + 1 + np.random.randn(100) * 5  # Hm s thc t (y ch l v d)

# Reshape X  ph hp vi Scikit-learn
X = X[:, np.newaxis]

# To cc m hnh v fit vi d liu hun luyn
degrees = np.arange(1, 10)
mse_list = []

plt.figure(figsize=(12, 6))
for i, degree in enumerate(degrees):
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    
    model = LinearRegression()
    model.fit(X_poly, y)
    
    # predict
    y_pred = model.predict(X_poly)
    
    # evaluate models using mean squared error
    mse = mean_squared_error(y, y_pred)
    mse_list.append(mse)
    
    # plot 
    plt.subplot(3, 3, i + 1)
    plt.scatter(X, y, s=10)
    plt.plot(X, y_pred, color='red')
    plt.title(f'Degree {degree}, MSE: {mse:.2f}')

plt.tight_layout()
plt.show()

# show MSE for every models
for degree, mse in zip(degrees, mse_list):
    print(f"Degree {degree}: MSE = {mse:.2f}")
```
