```python
import glob
import os
import pandas as pd

folders = glob.glob("LSTM*/*results.csv")

df = None
for fol in folders:
    if df is None:
        df = pd.read_csv(fol)
        df["run"] = df["run"] + 1
    else:
        df2 = pd.read_csv(fol)
        df2["run"] = df2["run"] + 1 + df["run"].max()
        df = pd.concat([df, df2], ignore_index=True)
    
df.to_csv("lstm_all.csv", index=False)
    
    
```
